```python
# Для датафрейма log из материалов занятия создайте столбец source_type по правилам:

# если источник traffic_source равен Yandex или Google, то в source_type ставится organic;
# для источников paid и email из России ставим ad;
# для источников paid и email не из России ставим other;
# все остальные варианты берём из traffic_source без изменений.
```

```python
import pandas as pd

log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
def table(row):
  if row['traffic_source'] == 'yandex':
    return 'organic'
  elif row['traffic_source'] == 'google':
    return 'organic'
  elif row['traffic_source']  == 'paid':
    if row['region'] == 'Russia':
      return 'ad'
  elif row['traffic_source'] == 'email' :
    if row['region'] == 'Russia':
      return 'ad'
    else:
       return 'other'
  else:
    return row['traffic_source']


```

```python
log['source_type'] = log.apply(table, axis = 1)
log.head(50)
```

```python
# Используйте файл с оценками фильмов ml-latest-small/ratings.csv.
# Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок.
# Под временем жизни понимается разница между максимальным и
# минимальным значениями столбца timestamp для данного значения userId.
```

```python
import pandas as pd

rating = pd.read_csv('ratings.csv')
rating.head()
```

```python
rating_count=rating.groupby('userId').count().reset_index()
rating_count=rating_count[['userId','rating']]
rating_count.head()
```

```python
count_100 = rating_count[rating_count['rating'] > 100]
count_100.rename(columns = {'rating' : 'rating_count'}, inplace=True)
count_100.head()
```

```python
timestamp_max = rating.groupby('userId').max().reset_index()
timestamp_max = timestamp_max[['userId', 'timestamp']]
timestamp_max.rename(columns = {'timestamp' : 'timestamp_max'}, inplace=True)
timestamp_max.head()
```

```python
timestamp_min = rating.groupby('userId').min().reset_index()
timestamp_min = timestamp_min[['userId', 'timestamp']]
timestamp_min.rename(columns = {'timestamp' : 'timestamp_min'}, inplace = True)
timestamp_min.head()
```

```python
avg_timestamp = count_100.merge(timestamp_max).merge(timestamp_min)
avg_timestamp.head()
```

```python
def average(row):
    mean = (row['timestamp_max'] - row['timestamp_min'])
    return mean
```

```python
avg_timestamp['average']=avg_timestamp.apply(average, axis=1)
avg_timestamp.head(20)
```

```python
#Дана статистика услуг перевозок клиентов компании по типам
#(см. файл “Python_13_join.ipynb” в разделе
#«Материалы для лекции “Продвинутый pandas”» ---- Ноутбуки к лекции «Продвинутый pandas»).
#Нужно сформировать две таблицы:
 #таблицу с тремя типами выручки для каждого client_id без указания адреса клиента;
 #аналогичную таблицу по типам выручки с указанием адреса клиента.
```

```python
import pandas as pd
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
pivot_table = rzd.merge(auto, how = 'outer').merge(air, how = 'outer')
pivot_table.loc[pivot_table.rzd_revenue.isnull(), 'rzd_revenue',]=0
pivot_table.loc[pivot_table.auto_revenue.isnull(), 'auto_revenue',]=0
pivot_table.loc[pivot_table.air_revenue.isnull(), 'air_revenue',]=0
pivot_table.head(20)
```

```python
pivot_table2 = pivot_table.merge(client_base, how='outer')
pivot_table2.head(10)
```
