```python
# Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

#оценка 2 и ниже — низкий рейтинг;
#оценка 4 и ниже — средний рейтинг;
#оценка 4.5 и 5 — высокий рейтинг.
#Результат классификации запишите в столбец class.


import pandas as pd

ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
def movies_class(param):
  """классифицирует фильмы из материалов занятия по следующим правилам:
        оценка 2 и меньше - низкий рейтинг
        оценка 4 и меньше - средний рейтинг
        оценка 4.5 и 5 - высокий рейтинг"""
  if param <= 2 :
    return 'низкий рейтинг'
  if param <= 4 :
    return 'средний рейтинг'
  if param > 4 or param <= 5:
    return 'высокий рейтинг'

movies['class'] = ratings['rating'].apply(movies_class)
movies.head()

```

```python
import keyword
from pandas.io.parsers.readers import read_csv
#Используйте файл keywords.csv.
#Нужно написать гео-классификатор, который каждой строке сможет выставить
#географическую принадлежность определённому региону.
#Т. е. если поисковый запрос содержит название города региона,
#то в столбце ‘region’ пишется название этого региона.
#Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

```

```python
def region_cities(param):
  """гео-классификатор, который каждой строке выставит географическую
     принадлежность определённому региону"""
  for region, cities in geo_data.items():
    for city in cities:
      if city in param:
         return region
  return 'undefined'

```

```python
keywords['region'] = keywords['keyword'].apply(region_cities)
keywords
```

```python
#Проверяем
cities = keywords[keywords['keyword'].isin(['москва ', 'ярославль', 'петербург', 'псков', 'мурманск', 'владивосток'])]
cities
```

```python
# Задание 3 (бонусное)

import numpy as np
import pandas as pd

years = np.arange(1950, 2011)

```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
movies.head()
```

```python
def production_year(movie):
  for i in years:
    if str(i) in movie:
      return i
  return 1900

```

```python
movies['year'] = movies['title'].apply(production_year)
movies
```

```python
# Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’
# отсортируйте результат по убыванию рейтинга.

ratings.head()
```

```python
movies_rating = movies.merge(ratings[['movieId', 'rating']])
movies_rating
```

```python
avg_rating = movies_rating.groupby('year').mean('rating').sort_values(by="rating", ascending=False)
avg_rating
```
